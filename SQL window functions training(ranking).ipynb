## SQL window functions training 
SQL dump from Yandex Practicum course is used

```python
import pandas as pd
%load_ext sql 
%sql postgresql://postgres:sqltest123@localhost/1
```

```python
%%sql result << 
SELECT user_id,
         event_dt,
       LAG(event_dt) OVER (PARTITION BY user_id ORDER BY event_dt) AS previous_order_dt,
       LEAD(event_dt) OVER (PARTITION BY user_id ORDER BY event_dt) AS next_order_dt
FROM online_store.orders
ORDER BY user_id; 
```

```python
#displaying results for ARPU SQL request
df = result.DataFrame()
display(df.head(11)) 
```

```python
%%sql result2 << 
SELECT *, 
       ROW_NUMBER() OVER (ORDER BY user_id)
FROM online_store.orders
ORDER BY user_id; 
```

```python
#displaying results for ARPPU SQL request
df2 = result2.DataFrame()
display(df2.head(11)) 
```

```python
%%sql result3 << 
SELECT *,
         NTILE(3) OVER (ORDER BY revenue)
FROM online_store.orders 
ORDER BY revenue; 
```

```python
df3 = result3.DataFrame()
display(df3.head(11)) 
```
